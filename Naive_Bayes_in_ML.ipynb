```python
# Naive Bayes(It classify data based on the probabilities of different classes given the features of the data)
import pandas as pd
import numpy as np
from sklearn.naive_bayes import MultinomialNB
```

```python
df = pd.read_csv('/content/golf_df(ML - Naive Bayes dataset).csv')
df
```

```python
print(df.info())
```

```python
# Below line iterates through each column (x) of the DataFrame (df) and converts the data type of that column to 'category'
df = df.apply(lambda x: x.astype('category'))

# Below line iterates through each column (x) of the DataFrame df
# cat.codes attribute returns a numerical representation of each category in the column. The resulting DataFrame with numerical codes is stored in df1
df1 = df.apply(lambda x: x.cat.codes)
df1

# This process maps each unique category (e.g., 'sunny', 'overcast', 'rainy' in 'Outlook') to a unique integer (e.g., 0, 1, 2)
```

```python
# train = df1[:10]: This line creates a new DataFrame called train by selecting the first 10 rows from df1. This will be your training data.
train = df1[:10]
train

# test = df1[-4:]: This line creates a new DataFrame called test by selecting the last 4 rows from df1. This will be your testing data.
test = df1[-4:]
test

# y_train = train.pop('Play'): This line selects the 'Play' column from the train DataFrame and removes it from train. The removed 'Play' column is stored in a new Series called y_train. This is the target variable for your training set.
y_train = train.pop('Play')
y_train

# x_train = train: This line assigns the remaining columns in the train DataFrame (which are now the features) to a new DataFrame called x_train. This is the feature set for your training data.
x_train = train
x_train

# y_test = test.pop('Play'): This line selects the 'Play' column from the test DataFrame and removes it. The removed 'Play' column is stored in a new Series called y_test.
y_test = test.pop('Play')
y_test

# x_test = test: This line assigns the remaining columns in the test DataFrame (the features) to a new DataFrame called x_test. This is the feature set for your testing data.
x_test = test
x_test
```

```python
model = MultinomialNB()
model
model_obj = model.fit(x_train, y_train)
model_obj
```

```python
y_out = model.predict(x_test)
y_out
```

```python
print('training accuracy: ', model.score(x_train, y_train))
print('testing accuracy: ', model.score(x_test, y_test))
```
