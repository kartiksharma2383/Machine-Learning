```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('/content/penguins(ML - Random Forest dataset).csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df.sex.unique()
```

```python
pd.get_dummies(df['sex']).head()
```

```python
sex = pd.get_dummies(df['sex'], drop_first = True)
sex.head()
```

```python
df.island.unique()
```

```python
pd.get_dummies(df['island']).head()
```

```python
island = pd.get_dummies(df['island'], drop_first = True)
island.head()
```

```python
# concatenate the above two dataframe to the original df
new_data = pd.concat([df, island, sex], axis = 1)
new_data.head()
```

```python
# Drop the repeated column
new_data.drop(['island', 'sex'], axis = 1, inplace = True)
new_data.head()
```

```python
new_data.drop(['Unnamed: 0'], axis = 1, inplace = True)
new_data.head()
```

```python
y = new_data.species
y.head()
```

```python
y.unique()
```

```python
# map function is used to convert categorical values into numeric values.
y = y.map({'Adelie':0, 'Chinstrap':1, 'Gentoo':2})
y.head()
```

```python
new_data.drop('species', inplace = True, axis = 1)
new_data.head()
```

```python
x = new_data
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
print('x_train', x_train.shape)
print('x_test', x_test.shape)
print('y_train', y_train.shape)
print('y_test', y_test.shape)
```

```python
# ensemble(It combine the predictions of multiple individual models (often called base estimators) to improve overall performance)
# using entropy criteria
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 5, criterion = 'entropy', random_state = 0)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report, accuracy_score
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
# using gini criteria
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 5, criterion = 'gini', random_state = 0)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```
